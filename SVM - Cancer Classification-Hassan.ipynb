# BREAST CANCER CLASSIFICATION USING SUPPORT VECTOR MACHINES


# STEP #1: PROBLEM STATEMENT


- Predicting if the cancer diagnosis is benign or malignant based on several observations/features 
- 30 features are used, examples:
        - radius (mean of distances from center to points on the perimeter)
        - texture (standard deviation of gray-scale values)
        - perimeter
        - area
        - smoothness (local variation in radius lengths)
        - compactness (perimeter^2 / area - 1.0)
        - concavity (severity of concave portions of the contour)
        - concave points (number of concave portions of the contour)
        - symmetry 
        - fractal dimension ("coastline approximation" - 1)

- Datasets are linearly separable using all 30 input features
- Number of Instances: 569
- Class Distribution: 212 Malignant, 357 Benign
- Target class:
         - Malignant
         - Benign




# DATA SOURCE: https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)

# STEP #4: MODEL TRAINING

```python
cancer_df
```

```python
X = cancer_df.drop(['target'], axis =1)
```

```python
X
```

```python
y = cancer_df['target']
```

```python
y
```

## Dividing the data into training and test by using Sci-kit Learn

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X ,y , test_size = 0.2, random_state =5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

# STEP #5: EVALUATING THE MODEL

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

# STEP #6: IMPROVING THE MODEL

```python
min_train = X_train.min()
```

```python
min_train
```

```python
range_train = X_train.max() - min_train
```

```python
range_train
```

```python
X_train_scaled = (X_train - min_train)/range_train
```

```python
X_train_scaled
```

```python
sns.scatterplot(x=X_train['mean area'], y=X_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x=X_train_scaled['mean area'], y=X_train['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
```

```python
min_test
```

```python
range_test = (X_test - min_test).max()
```

```python
range_test
```

```python
X_test_scaled = (X_test - min_test)/range_test
```

```python
X_test_scaled
```

```python
sns.scatterplot(x=X_test['mean area'], y=X_test['mean smoothness'], hue = y_test)
```

```python
sns.scatterplot(x=X_test_scaled['mean area'], y=X_test_scaled['mean smoothness'], hue = y_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True, fmt='d')
```

```python
print(classification_report(y_test, y_predict))
```
